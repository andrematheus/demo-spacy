# "Parts-of-speech" e "lemmas" com spaCY

spaCY oferece "parts-of-speech" (nome, verbo, advérbio, etc.) e "lemmas" -– variantes padronizadas de grupos de palavras relacionadas (por exemplo, o lemma de "escreveu" e "escreve" é "escrever").

```python
import spacy
from utils import *
```

```python
nlp = spacy.load('en')
doc = nlp('The carnivorous cactuses ate your hambuger.')

headers = ['token', 'pos_', 'tag_']
rows = []
for token in doc:
    rows.append([token, token.pos_, token.tag_])
    
table(headers, rows)
```
